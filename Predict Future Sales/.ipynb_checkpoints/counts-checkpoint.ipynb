```python
from collections import Counter
import pandas as pd
```

```python
#Возвращает название товара по заданному DF с товарами, названию столбца с идентификаторами и по заданному идентификатору
def get_item_name_by_id(items_df, items_column_id_name, ident):
    return items_df[items_df[items_column_id_name] == ident]
```

```python
sales_fname = 'sales_train_v2.csv'
sales_fname_data = 'sales_train_data.csv'
sales_column_names = ['Date', 'DateBlockNum', 'ShopId', 'ItemId', 'ItemPrice', 'ItemCntDay']
items_fname = 'items.csv'
items_column_names = ['ItemName', 'ItemId', 'ItemCategoryId']
```

```python
with open(sales_fname_data, 'r') as f:
    for i in range(0):
        f.readline()
    for i in range(7):
        x = f.readline()
        print(x, '\n')
   
```

```python
sales = pd.read_csv(sales_fname_data, engine='c',
                  header = 0, names = sales_column_names)
items = pd.read_csv(items_fname, engine='c',
                  header = 0, names = items_column_names)
```

```python
items['ItemName'] = items['ItemName'].astype(str)
```
