```python
from collections import Counter
import pandas as pd
```

```python
def prepare_data(data, y_col_name, test_size = 0.15):
    
    import datetime as dt
    # считаем индекс в датафрейме, после которого начинается тестовыый отрезок
    test_index = int(data.index[int((1-test_size)*len(data))])
    
    data = data.dropna()
    
    X_train = data.loc[:test_index-1].drop([y_col_name], axis=1)
    y_train = data.loc[:test_index-1][y_col_name]
    X_test = data.loc[test_index:].drop([y_col_name], axis=1)
    y_test = data.loc[test_index:][y_col_name]

    return X_train, X_test, y_train, y_test
```

```python
def plot_data(header, y_train, y_test, prediction_train, prediction_test):
    from sklearn.metrics import mean_absolute_error, explained_variance_score, r2_score
    import matplotlib.pyplot as plt

    plt.figure(figsize=(15, 7))
    plt.plot(y_train, label="actual train")
    plt.plot(y_test, label="actual test")
    plt.plot(prediction_train, label = "prediction train")
    plt.plot(prediction_test, label="prediction test")
    plt.legend(loc="best")
    plt.title(header + "\n Metrics on test part: \n Mean absolute error {}".format(mean_absolute_error(prediction_test, y_test)) 
              + "\n Explained variance regression score {}".format(explained_variance_score(prediction_test, y_test))
              + "\n R2 score {}".format(r2_score(prediction_test, y_test)))
    plt.grid(True);
```

```python
def plot_model(model, model_name, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    prediction_test = model.predict(X_test)
    prediction_test = list(map(round, prediction_test))
    prediction_test = list(map(int, prediction_test))
    
    prediction_train = model.predict(X_train)
    prediction_train = list(map(round, prediction_train))
    prediction_train = list(map(int, prediction_train))
    
    print(prediction_test)
    
    pred_test = pd.DataFrame(prediction_test, index = y_test.index)
    pred_train = pd.DataFrame(prediction_train, index = y_train.index)
    #print(pred_test)
    
    plot_data(model_name, y_train, y_test, pred_train, pred_test)
```

```python
def get_pred_by_one_model(model, X_train, X_test, Y_train):
    model.fit(X_train, y_train)
    prediction_train = model.predict(X_train)
    prediction_train = list(map(round, prediction_train))
    prediction_train = list(map(int, prediction_train))
    return prediction_train
```

```python
def get_pred_value_by_two_models(model1, model2, X_train, X_test, Y_train):
    p1 = getpred_by_one_model(model1, X_train, X_test, Y_train)
    p2 = getpred_by_one_model(model2, X_train, X_test, Y_train)
    return int(round((p1[0] + p2[0])/2))
```
